```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/all_examples.csv', delimiter='\t')
```

```python
df.head()
```

```python
df1 = pd.read_csv('data/cate2guidelines.csv', delimiter='\t')
```

```python
df1.head()
```

```python
dataset_counts = df['dataset'].value_counts()
```

```python
dataset_counts
```

```python
plt.figure(figsize=(8, 8))
plt.pie(dataset_counts, labels=dataset_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')

plt.title('Distribution of Hate Speech Examples by Dataset')
plt.show()
```

```python
category_counts = df['category'].value_counts()
```

```python
category_counts
```

```python
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')

plt.xlabel('Category')
plt.ylabel('Number of Hate Speech Examples')
plt.title('Distribution of Hate Speech Examples by Category')
plt.show()
```

```python
df1.rename(columns={'guideline': 'category'}, inplace=True)
```

```python
merged_df = pd.merge(df, df1, on='category', how='left')
```

```python
merged_df.head()
```

```python
tier_counts = merged_df['tier'].value_counts()
```

```python
tier_counts
```

```python
plt.figure(figsize=(10, 6))
tier_counts.plot(kind='bar')

plt.xlabel('Tier')
plt.ylabel('Number of Hate Speech Examples')
plt.title('Distribution of Hate Speech Examples by Tier')
plt.show()
```
